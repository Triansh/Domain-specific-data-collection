```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_web = pd.read_csv('../phase_1_data_collection/combined_5k_webmineral_clean.csv')
df_wiki = pd.read_csv('../phase_1_data_collection/combined_wiki_mineral_clean.csv')
df_ima = pd.read_csv('../phase_1_data_collection/ima_mineral_list.csv', index_col=0)
```

```python
df_wiki.head(2)
```

```python
len(df_wiki.columns)
```

```python
row = [df_wiki.shape[0], df_web.shape[0], df_ima.shape[0]]
col = [df_wiki.shape[1], df_web.shape[1], df_ima.shape[1]]
name = ['Wikipedia', 'Web Mineral', 'IMA']
```

```python
df_plot = pd.DataFrame(list(zip(row, col, name)),
                       columns=['row', 'col', 'name'])
```

```python
df_plot
```

```python
plt.rcParams["figure.figsize"] = [15, 8]
sns.set_style('whitegrid')
sns.set(font_scale=1.3)
sns.barplot(x="name", y="row", data=df_plot, hue="name")

plt.ylabel("Number of Minerals", fontsize=15)
plt.xlabel("Sources", fontsize=15)
plt.title("Minerals scrapped from various sources")
plt.legend(title="Sources", fontsize=15, title_fontsize=15)
```

```python
plt.rcParams["figure.figsize"] = [15, 8]
sns.set_style('whitegrid')
sns.set(font_scale=1.3)
sns.barplot(x="name", y="col", data=df_plot, hue="name")

plt.ylabel("Number of Attributes", fontsize=15)
plt.xlabel("Sources", fontsize=15)
plt.title("Attributes found per source")
plt.legend(title="Sources", fontsize=15, title_fontsize=15)
```
