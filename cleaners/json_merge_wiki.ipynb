```python
import os
import json
import pandas as pd
import numpy as np
from pprint import pprint
from copy import deepcopy
```

```python
global_min_dict = {}
```

```python
for file in os.listdir('../data/wikipedia'):
    if file.endswith('json') and file.startswith('minerals'):
        print(file)
        with open(f'../data/wikipedia/{file}', 'r') as f:
            data = json.load(f)
            global_min_dict.update(data)
```

```python
print(len(global_min_dict))
attr = [len(v) for v in global_min_dict.values()]
print(np.median(attr))
```

```python
with open('../data/wikipedia/combined-wiki-all.json', 'w') as f:
    json.dump(global_min_dict, f)
```

```python
copy_mineral = deepcopy(global_min_dict)
```

```python
print(len(copy_mineral))
attr = [len(v) for v in copy_mineral.values()]
print(np.median(attr))
print(np.mean(attr))
```

```python
records = []
for k, v in copy_mineral.items():
    temp = deepcopy(v)
    temp['Mineral name'] = k
    records.append(temp)
```

```python
df = pd.DataFrame.from_records(records)
df=df.set_index('Mineral name')
df.head(3)
```

```python
copy_df = df.copy()
```

```python
row_remove_perc = 1.0
min_count = int((row_remove_perc / 100) * copy_df.shape[1] + 1)
copy_df.dropna(axis=0, how='any', thresh=min_count, inplace=True)
copy_df.reset_index(drop=True, inplace=True)
```

```python
copy_df.info()
```

```python
col_remove_perc = 30.0
min_count = int((col_remove_perc / 100) * copy_df.shape[0] + 1)
copy_df.dropna(axis=1, how='any', thresh=min_count, inplace=True)
```
