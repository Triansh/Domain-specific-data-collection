```python
minrels_dic2 = {}
dic2_set = set()
```

```python
#line_re = re.compile(u"(a href=\")(\/wiki\/.*?)(\".*title\=\")(.*?)(\">)",re.M)
line_re = re.compile(u"(^<li><a href=\")(\/wiki\/.*?)(\".*title\=\")(.*?)(\">)", re.M)

f = open(
    "C:\Arvin\Personal\IIIT\IIIT_21-22_Mansoon IR & IE\Home Work\Major Projects\wiki_minerals.txt",
    "r")
txt = f.readlines()
```

```python
for i in range(len(txt)):
    line_dtls = re.finditer(line_re, txt[i])
    for line_dtl in line_dtls:
        link = line_dtl.group(2).lower()
        title = line_dtl.group(4).lower()
        print(link + " - " + title)

        minrels_dic2[title] = "https://en.wikipedia.org" + link
        dic2_set.add(title)
```

```python
set1 = set()
set2 = set()
for e in dic1_set:
    set1.add(e.lower())

for e in dic2_set:
    set2.add(e.lower())

url_f = open(
    "C:\Arvin\Personal\IIIT\IIIT_21-22_Mansoon IR & IE\Home Work\Major Projects\wiki_minerals_url.txt",
    "w", encoding="utf-8")
url_no_match = open(
    "C:\Arvin\Personal\IIIT\IIIT_21-22_Mansoon IR & IE\Home Work\Major Projects\wiki_minerals_url_no_match.txt",
    "w", encoding="utf-8")
url_mising1 = open(
    "C:\Arvin\Personal\IIIT\IIIT_21-22_Mansoon IR & IE\Home Work\Major Projects\wiki_minerals_url_mising1.txt",
    "w", encoding="utf-8")
url_mising2 = open(
    "C:\Arvin\Personal\IIIT\IIIT_21-22_Mansoon IR & IE\Home Work\Major Projects\wiki_minerals_url_mising2.txt",
    "w", encoding="utf-8")
```

```python
for m in minrels_set:
    mineral = m.lower()
    if mineral in set1 and mineral in set2:
        if minrels_dic1[mineral] == minrels_dic2[mineral]:
            line = mineral + "||" + minrels_dic1[mineral] + "\n"
            url_f.write(line)

    if mineral in set1 and mineral in set2:
        if minrels_dic1[mineral] != minrels_dic2[mineral]:
            line = mineral + "||" + minrels_dic1[mineral] + "||" + minrels_dic2[mineral] + "\n"
            url_no_match.write(line)

    if mineral in set2 and mineral not in set1:
        line = mineral + "||" + minrels_dic2[mineral] + "\n"
        url_mising1.write(line)

    if mineral in set1 and mineral not in set2:
        line = mineral + "||" + minrels_dic1[mineral] + "\n"
        url_mising2.write(line)

url_f.close()
url_no_match.close()
url_mising1.close()
url_mising1.close()
```
