```python
def get_atr_val(mnrl_webdat,mnrl_wiki,atr):
    #m=''
    value =''
    for i in attr_dic[atr].keys():
        lst = attr_dic[atr][i]
        
        if lst[0] == 'webdat':
            try:
                value = webdat_dic[mnrl_webdat][lst[1]]
                #m = mnrl_webdat
                break
            except:
                a=1
        else:
            try:
                value = wiki_dic[mnrl_wiki][lst[1]]
                #m = mnrl_wiki
                break
            except:
                a=1
    return value
        
```

```python
merged_dic  = defaultdict(lambda:defaultdict(str))

for mnrl in all_mnrl:
    mName_webdat = ''
    mName_wiki = ''
    for m in webdat_dic.keys():
        if mnrl.lower() == m.lower():
            mName_webdat = m
            break
    
    for m in  wiki_dic.keys():
        if mnrl.lower() == m.lower():
            mName_wiki = m
            break
    
    if mName_wiki != '':
        m= mName_wiki
        
    if mName_webdat != '':
        m= mName_webdat
        
    for atr in attr_dic.keys():
        val = get_atr_val(mName_webdat,mName_wiki, atr)
        if m != '':
            merged_dic[m][atr] = val
    
        
        
```
